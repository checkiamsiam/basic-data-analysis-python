```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read The CSV File
sales_csv_path = "/content/dataset/Sales.csv"
data = pd.read_csv(sales_csv_path)
# Format Data
data_as_array_of_objects = data.to_dict(orient='records')
```

```python
# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(data_as_array_of_objects)
```

In the given csv file there are very ammount of random data to who

```python
# Line chart for total sales by brand
plt.figure(figsize=(10, 6))
brand_sales = df.groupby('Brands')['Selling Price'].sum().sort_values(ascending=False)
brand_sales.plot(kind='line', marker='o', color='b')
plt.title('Total Sales by Brand')
plt.xlabel('Brand')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart for popular colors
plt.figure(figsize=(8, 8))
color_count = df['Colors'].value_counts()
color_count.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Popular Colors')
plt.ylabel('')
plt.show()
```

```python
# Scatter plot for most demanded memory sizes
plt.figure(figsize=(10, 6))
memory_demand = df['Memory'].value_counts()
plt.scatter(memory_demand.index, memory_demand.values, color='g', s=100)
plt.title('Most Demanded Memory Sizes')
plt.xlabel('Memory Size')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Box plot for average rating by brand
plt.figure(figsize=(10, 6))
df.boxplot(column='Rating', by='Brands', grid=False)
plt.title('Average Rating by Brand')
plt.suptitle('')  # Suppress the default title to only show the specified title
plt.xlabel('Brand')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram for Selling Price
plt.figure(figsize=(10, 6))
plt.hist(df['Selling Price'], bins=10, edgecolor='black', color='c')
plt.title('Distribution of Selling Prices')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
